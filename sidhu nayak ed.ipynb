```python
import pandas as pd

# Load the dataset
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")

# Check dataset size
df.shape

# Preview first 5 rows
df.head()
```

```python
# Check column names, data types, and non-null counts
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Count duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Check new dataset size
df.shape
```

```python
# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Display cleaned column names
df.columns
```

```python
# Convert all columns to integer type
df = df.astype(int)

# Final check of dataset
df.info()
```

```python
# Shows basic statistics like mean, min, max, etc.
df.describe()
```

```python
import matplotlib.pyplot as plt

# Distribution of diabetes classes
df["Diabetes_012"].value_counts().plot(kind="bar")
plt.xlabel("Diabetes Class (0 = No, 1 = Pre, 2 = Yes)")
plt.ylabel("Count")
plt.title("Distribution of Diabetes Classes")
plt.show()
```

```python
# Distribution of BMI values
df["BMI"].plot(kind="hist", bins=30)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("BMI Distribution")
plt.show()
```

```python
# Distribution of age groups
df["Age"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Group Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr = df.corr()

# Create figure
plt.figure(figsize=(12, 10))

# Plot heatmap
plt.imshow(corr, cmap="viridis")
plt.colorbar()

# Set column names on axes
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

# Add title
plt.title("Correlation Heatmap")

# Annotate correlation values in each cell
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, round(corr.iloc[i, j], 2),
                 ha="center", va="center", fontsize=6)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a box plot to detect outliers in BMI
plt.figure(figsize=(6, 4))
plt.boxplot(df["BMI"])
plt.title("Box Plot of BMI (Outlier Detection)")
plt.ylabel("BMI")
plt.show()
```
